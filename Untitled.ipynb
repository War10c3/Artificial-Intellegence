```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.countplot(x='bedrooms', data=df, palette='RdBu')
plt.xlabel('bedrooms (bedrooms)')
plt.show()
```

```python
#copying the original dataframe to 'temp_df'.
temp_df = df[['id','price','condition', 'sqft_lot15', 'bedrooms','grade']].copy()

#Adding new features to dataframe.
temp_df["Sentiments"] = temp_df["condition"].apply(lambda condition: "positive" if condition > 3 else \
                                              ("negative" if condition < 3 else "not defined"))
temp_df["Usefulness"] = (temp_df["sqft_lot15"]/temp_df["price"]).apply\
(lambda n: ">75%" if n > 0.75 else ("<25%" if n < 0.25 else ("25-75%" if n >= 0.25 and\
                                                                        n <= 0.75 else "useless")))

temp_df.loc[temp_df.price == 0, 'Usefulness'] = ["useless"]
# Removing all rows where 'Score' is equal to 3
#temp_df = temp_df[temp_df.Score != 3]
#Lets now observe the shape of our new dataframe.
temp_df.shape
```

```python
temp_df.info()
```

```python
sns.countplot(x='Sentiments', order=["positive", "negative"], data=temp_df, palette='RdBu')
plt.xlabel('Sentiments')
plt.show()
temp_df.Sentiments.value_counts()
```

```python
pos = temp_df.loc[temp_df['Sentiments'] == 'positive']
pos = pos[0:25000]

neg = temp_df.loc[temp_df['Sentiments'] == 'negative']
neg = neg[0:25000]
```

```python
temp_df.Usefulness.value_counts()
```

```python
sns.countplot(x='Usefulness', order=['useless', '>75%', '25-75%', '<25%'], data=temp_df, palette='RdBu')
plt.xlabel('Usefulness')
plt.show()
```

```python
sns.violinplot(x='Usefulness', y='price', order=[">75%", "<25%"], \
               data=temp_df, palette='RdBu')
plt.ylim(-50, 400)
plt.show()
```
